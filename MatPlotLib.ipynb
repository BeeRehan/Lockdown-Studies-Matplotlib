# Excersise

```python
x = [1,2,3,4]
y = [10, 25, 35, 40]
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,y)
ax.set(title="Sample",xlabel="x",ylabel="y")
fig.savefig('saves/mat1')
```

# Scatter and Bar

```python
n1 = np.linspace(0,10,75)

fig, ax = plt.subplots()
ax.scatter(n1, n1**3);
ax.set(title="Sample-Scatter",xlabel="x-axis",ylabel="y-axis")
fig.savefig('saves/mat2sactter')
```

```python
d1 = {'Yasin':50,
       'Lara':75,
     'L2R':100}
fig, ax = plt.subplots()
ax.bar(d1.keys(),d1.values());
ax.set(title='love-counter',ylabel='% percentage',xlabel='Person');
fig.savefig('saves/mat3bar')
```

```python
fig, ax = plt.subplots()
ax.barh(list(d1.keys()),list(d1.values()));
ax.set(title='love-counter',xlabel='% percentage',ylabel='Person');
fig.savefig('saves/mat3barh')
```

# Histogram

```python
fig, ((ax1,ax2), (ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax1.plot(x,y);
ax2.scatter(x,y);
ax3.bar(d1.keys(),d1.values());
ax4.hist(np.random.randn(1000))
fig.savefig('saves/mat4mixed')
```
